```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../outnow_3.csv').drop(columns="Unnamed: 0")
```

```python
df.head()
```

```python
df_comp = df[df['COMP'] == 1]
df_comp_num = df_comp.select_dtypes('number').drop(columns=["affiliation_id", "cited_by_count"]).iloc[:,27:]
df_comp_num
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df_comp_num)
```

```python
pca = PCA(n_components=3)
data_pca = pca.fit_transform(data_scaled)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
```

```python
ax.scatter(data_pca[:, 0], data_pca[:, 1], data_pca[:, 2], s=50, alpha=0.7, c='blue')
```

```python
ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
ax.set_title('3D PCA Scatter Plot of Vectors')

```

```python
plt.show()
```
